```python
import torch
import os
import pandas as pd
import numpy as np
from PIL import Image
import cv2
import seaborn as sns
from tqdm import tqdm
from torchvision import datasets
import torchvision.transforms as T
from torch.utils.data import DataLoader, Dataset
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import matplotlib.pyplot as plt
from torchsummary import summary
import time
```

```python
class BaseCNN(nn.Module):
    def __init__(self):
        super(BaseCNN, self).__init__()
        self.features = nn.Sequential(
            nn.Conv2d(3, 64, kernel_size=3, padding='same'),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2),
            nn.Dropout2d(p=0.2),
            
            nn.Conv2d(64, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2),
            nn.Dropout2d(p=0.2),
            
            nn.Conv2d(128, 128, kernel_size=3, padding='same'),
            nn.BatchNorm2d(128),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2),
            nn.Dropout2d(p=0.2),
            
            nn.Conv2d(128, 256, kernel_size=3, padding='same'),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(2, stride=2),
            nn.Dropout2d(p=0.2),
            
            nn.Conv2d(256, 256, kernel_size=3, padding='same'),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2),
            nn.Dropout2d(p=0.2),

            nn.Conv2d(256, 256, kernel_size=3, padding='same'),
            nn.BatchNorm2d(256),
            nn.ReLU(inplace=True),
            nn.MaxPool2d(3, stride=2),
            nn.Dropout2d(p=0.2),
            nn.Flatten()
        )
    
    def forward(self, x):
        return self.features(x)

class AgeEstimator(nn.Module):
    def __init__(self):
        super(AgeEstimator, self).__init__()
        self.base_cnn = BaseCNN()
        self.classifier = nn.Sequential(
            nn.Linear(256*2*2, 128),
            nn.BatchNorm1d(128),
            nn.Dropout(p=0.4),
            nn.ReLU(inplace=True),
            
            nn.Linear(128, 64),
            nn.BatchNorm1d(64),
            nn.Dropout(p=0.4),
            nn.ReLU(inplace=True),
            
            nn.Linear(64, 32),
            nn.BatchNorm1d(32),
            nn.Dropout(p=0.4),
            nn.ReLU(inplace=True),
            
            nn.Linear(32, 1)
        )
    
    def forward(self, x):
        x = self.base_cnn(x)
        x = self.classifier(x)
        return x

class GenderEstimator(nn.Module):
    def __init__(self):
        super(GenderEstimator, self).__init__()
        self.base_cnn = BaseCNN()
        self.classifier = nn.Sequential(
            nn.Linear(256*2*2, 64),
            nn.BatchNorm1d(64),
            nn.Dropout(p=0.4),
            nn.ReLU(inplace=True),
            
            nn.Linear(64, 32),
            nn.BatchNorm1d(32),
            nn.Dropout(p=0.4),
            nn.ReLU(inplace=True),

            nn.Linear(32, 2),
            nn.Softmax(dim=1)
        )
    
    def forward(self, x):
        x = self.base_cnn(x)
        x = self.classifier(x)
        return x
```

```python
class Data(Dataset):
    def __init__(self, dir, transform):
        self.dir = dir
        self.transform = transform
        self.images = os.listdir(dir)
        
    def __len__(self):
        return len(self.images)
        
    def __getitem__(self, index):
        img_path = os.path.join(self.dir, self.images[index])
        img = cv2.imread(img_path)
        img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
        filename = self.images[index]
        labels = filename.split('_')
        age = torch.Tensor([int(labels[0])])
        gender = torch.Tensor([int(labels[1])])
        if self.transform:
            img = self.transform(img)
        return img, age, gender
    
```

```python
EPOCHS = 50
IMG_HEIGHT = 180
IMG_WIDTH = 180
NUM_WORKERS = 2
DEVICE = 'cuda' if torch.cuda.is_available() else 'cpu'
BATCH_SIZE = 32
PIN_MEMORY=True
DIR =r'D:\Codes\MLDS\Age and Gender Estimator\archive\UTKFace'
LR = 1e-3
```

```python
def train(loader, age_model, gender_model, age_optimizer, gender_optimizer, age_loss_fn, gender_loss_fn, device):
    age_model.train()
    gender_model.train()
    
    loop = tqdm(enumerate(loader), total=len(loader))
    for batch_idx, (data, age_target, gender_target) in loop:
        data, age_target, gender_target = data.to(device), age_target.to(device), gender_target.to(device)
        
        age_output = age_model(data)
        age_loss = age_loss_fn(age_output, age_target)

        gender_output = gender_model(data)
        gender_target = gender_target.view(-1).long()
        gender_loss = gender_loss_fn(gender_output, gender_target)

        age_optimizer.zero_grad()
        gender_optimizer.zero_grad()

        age_loss.backward()
        age_optimizer.step()

        gender_loss.backward()
        gender_optimizer.step()
        
        if batch_idx % 10 == 0:
            print(f"Train Batch: {batch_idx} \t Age Loss: {age_loss.item():.6f} \t Gender Loss: {gender_loss.item():.6f}")
```

```python
transforms = T.Compose([T.ToTensor(),
    T.Resize((IMG_HEIGHT, IMG_WIDTH))
])
age_model = AgeEstimator().to(DEVICE)
gender_model = GenderEstimator().to(DEVICE)
age_loss_fn = nn.MSELoss()
gender_loss_fn = nn.CrossEntropyLoss()
age_optimizer = optim.Adam(age_model.parameters(), lr = LR)
gender_optimizer = optim.Adam(gender_model.parameters(), lr = LR)
```

```python
summary(gender_model, input_size = (3, 180, 180))
```

```python
dataset = Data(DIR, transform = transforms)
train_size = int(0.8 * len(dataset))
val_size = int(0.1 * len(dataset))
test_size = len(dataset) - train_size - val_size

train_ds, val_ds, test_ds = torch.utils.data.random_split(
    dataset, [train_size, val_size, test_size])

```

```python
train_ds = DataLoader(train_ds,batch_size=BATCH_SIZE,shuffle=True)
val_ds = DataLoader(val_ds,batch_size=BATCH_SIZE,shuffle=False)   
test_ds = DataLoader(test_ds,batch_size=BATCH_SIZE,shuffle=False)
```

```python
def test(loader, age_model, gender_model, age_loss_fn, gender_loss_fn, device='cuda' if torch.cuda.is_available() else 'cpu'):
    age_test_loss = 0
    age_mae = 0
    gender_test_loss = 0
    gender_correct = 0
    total_samples = 0
    
    age_model.eval()
    gender_model.eval()
    loop = tqdm(loader, total=len(loader))
    
    with torch.no_grad():
        for (data, age_target, gender_target) in loop:
            data, age_target, gender_target = data.to(device), age_target.to(device), gender_target.to(device)
            
            age_output = age_model(data)
            gender_output = gender_model(data)
            gender_target = gender_target.view(-1).long()
            
            age_test_loss += age_loss_fn(age_output, age_target).item()
            age_mae += torch.abs(age_output - age_target).sum().item()
            
            gender_loss = gender_loss_fn(gender_output, gender_target.long()).item()
            gender_test_loss += gender_loss
            _, gender_pred = torch.max(gender_output, 1)
            gender_correct += (gender_pred == gender_target).sum().item()
            
            total_samples += data.size(0)

    age_test_loss /= total_samples
    age_mae /= total_samples
    age_rmse = np.sqrt(age_test_loss)
    
    gender_test_loss /= total_samples 
    gender_accuracy = gender_correct / total_samples

    print(f'Age - Test Loss: {age_test_loss:.4f} \t MAE: {age_mae:.4f} \t RMSE: {age_rmse:.4f}')
    print(f'Gender - Test Loss: {gender_test_loss:.4f} \t Accuracy: {gender_accuracy:.4f}')
```

```python
for epoch in range(EPOCHS):
    train(train_ds, age_model, gender_model, age_optimizer, gender_optimizer, age_loss_fn, gender_loss_fn, DEVICE)
```
